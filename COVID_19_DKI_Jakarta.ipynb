```python
import numpy as np
import pandas as pd
import requests
resp_jakarta = requests.get('https://data.covid19.go.id/public/api/prov_detail_DKI_JAKARTA.json', verify=False)
cov_jakarta_raw = resp_jakarta.json()
cov_jakarta = pd.DataFrame(cov_jakarta_raw['list_perkembangan'])


cov_jakarta_tidy = (cov_jakarta.drop(columns=[item for item in cov_jakarta.columns 
                                               if item.startswith('AKUMULASI') 
                                                  or item.startswith('DIRAWAT')])
                           .rename(columns=str.lower)
                           .rename(columns={'kasus': 'kasus_baru'})
                  )
cov_jakarta_tidy['tanggal'] = pd.to_datetime(cov_jakarta_tidy['tanggal']*1e6, unit='ns')

cov_jakarta_pekanan = (cov_jakarta_tidy.set_index('tanggal')['kasus_baru']
                                   .resample('W')
                                   .sum()
                                   .reset_index()
                                   .rename(columns={'kasus_baru': 'jumlah'})
                    )
cov_jakarta_pekanan['tahun'] = cov_jakarta_pekanan['tanggal'].apply(lambda x: x.year)
cov_jakarta_pekanan['pekan_ke'] =cov_jakarta_pekanan['tanggal'].apply(lambda x:x.weekofyear)
cov_jakarta_pekanan = cov_jakarta_pekanan[['tahun', 'pekan_ke', 'jumlah']]

cov_jakarta_pekanan['jumlah_pekanlalu'] = cov_jakarta_pekanan['jumlah'].shift().replace(np.nan, 0).astype(np.int)
cov_jakarta_pekanan['lebih_baik'] = cov_jakarta_pekanan['jumlah'] < cov_jakarta_pekanan['jumlah_pekanlalu']

import matplotlib.pyplot as plt

plt.clf()
jml_tahun_terjadi_covid19 = cov_jakarta_pekanan['tahun'].nunique()
tahun_terjadi_covid19 = cov_jakarta_pekanan['tahun'].unique()
fig, axes = plt.subplots(nrows=jml_tahun_terjadi_covid19,
             figsize=(10,3*jml_tahun_terjadi_covid19))

fig.suptitle('Kasus Pekanan Positif COVID-19 di DKI Jakarta', 
       y=1.00, fontsize=16, fontweight='bold', ha='center')
for i, ax in enumerate(axes):
  ax.bar(data=cov_jakarta_pekanan.loc[cov_jakarta_pekanan['tahun']==tahun_terjadi_covid19[i]],
       x='pekan_ke', height='jumlah', 
       color=['mediumseagreen' if x is True else 'salmon' 
          for x in cov_jakarta_pekanan['lebih_baik']])
  if i == 0:
    ax.set_title('Kolom hijau menunjukkan penambahan kasus baru lebih sedikit dibandingkan satu pekan sebelumnya', 
           fontsize=10)
  elif i == jml_tahun_terjadi_covid19-1:
    ax.text(1, -0.2, 'Sumber data: covid.19.go.id', color='blue', 
      ha='right', transform=ax.transAxes)
    
  ax.set_xlim([0, 52.5])
  ax.set_ylim([0, max(cov_jakarta_pekanan['jumlah'])])
  ax.set_xlabel('')
  ax.set_ylabel('Jumlah kasus %d'%(tahun_terjadi_covid19[i],))
  ax.grid(axis='y')
          
plt.tight_layout()
plt.show()
```
